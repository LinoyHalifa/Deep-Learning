```python
# Define the path to the dataset
dataset_path = r"C:/Users/linoyh/ScientificPython/DL/Pictures"

# Function to check for corrupted images
def check_corrupted_images(category):
    folder = os.path.join(dataset_path, category)
    corrupted = []
    
    # Loop through each image in the folder
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)
        try:
            # Try to open and verify the image
            img = Image.open(img_path)
            img.verify()  # If there's an error, the image is corrupted
        except (IOError, SyntaxError):
            corrupted.append(img_name)
    
    return corrupted

# Check for corrupted images in "Dog Image" and "Cat Images" folders
corrupted_dogs = check_corrupted_images("Dog Image")
corrupted_cats = check_corrupted_images("Cat Images")

# Print the list of corrupted images
print(f"Corrupted dog images: {corrupted_dogs}")
print(f"Corrupted cat images: {corrupted_cats}")
```

```python
#4

# Define paths
dataset_path = r"C:/Users/linoyh/ScientificPython/DL/Pictures"
output_path = r"C:/Users/linoyh/ScientificPython/DL/Split_Dataset"

# Define split ratios
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# Categories
categories = ["Dog Image", "Cat Images"]

# Create directories for train, validation, and test sets
for split in ["train", "validation", "test"]:
    for category in categories:
        os.makedirs(os.path.join(output_path, split, category), exist_ok=True)

# Function to split dataset
def split_data(category):
    category_path = os.path.join(dataset_path, category)
    images = os.listdir(category_path)
    random.shuffle(images)  # Shuffle images

    # Calculate split indices
    train_idx = int(len(images) * train_ratio)
    val_idx = train_idx + int(len(images) * val_ratio)

    # Split images
    train_files = images[:train_idx]
    val_files = images[train_idx:val_idx]
    test_files = images[val_idx:]

    # Move files to respective folders
    for img_name in train_files:
        shutil.copy(os.path.join(category_path, img_name), os.path.join(output_path, "train", category))

    for img_name in val_files:
        shutil.copy(os.path.join(category_path, img_name), os.path.join(output_path, "validation", category))

    for img_name in test_files:
        shutil.copy(os.path.join(category_path, img_name), os.path.join(output_path, "test", category))

# Apply function to both categories
for category in categories:
    split_data(category)

print("Dataset split completed successfully!")
```

```python
#9 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import pandas as pd




def train_cnn(conv_layers=3, learning_rate=0.001, epochs=10):
    """Function to create and train CNN model with given hyperparameters."""
    
    model = Sequential()
    
    # Add convolutional layers based on the selected parameter
for _ in range(conv_layers):
    model.add(Conv2D(32 * (_ + 1), (3,3), activation='relu', input_shape=(64, 64, 1)))
    model.add(MaxPooling2D(2,2))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))  

# Define the optimizer
optimizer = Adam(learning_rate=learning_rate)

# Compile the model
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(train_generator, validation_data=val_generator, epochs=epochs, verbose=1)

# Evaluate performance
test_loss, test_acc = model.evaluate(test_generator, verbose=1)

print(f"🔹 Results for conv_layers={conv_layers}, learning_rate={learning_rate}, epochs={epochs}:")
print(f"Test Accuracy: {test_acc:.4f}")

return test_acc
```

```python
results = []

# Combination 1: 2 layers, learning rate 0.1, 10 epochs
results.append(("2 Layers, LR=0.1, 10 Epochs", train_cnn(conv_layers=2, learning_rate=0.1, epochs=10)))

# Combination 2: 3 layers, learning rate 0.01, 20 epochs
results.append(("3 Layers, LR=0.01, 20 Epochs", train_cnn(conv_layers=3, learning_rate=0.01, epochs=20)))

# Combination 3: 4 layers, learning rate 0.001, 30 epochs
results.append(("4 Layers, LR=0.001, 30 Epochs", train_cnn(conv_layers=4, learning_rate=0.001, epochs=30)))

# Create results table
results_df = pd.DataFrame(results, columns=["Configuration", "Test Accuracy"])

# Display the results
# tools.display_dataframe_to_user(name="Hyperparameter Tuning Results", dataframe=results_df)
display(results_df)  # In Jupyter, this will display a nicely formatted table

```

```python
import cv2
import numpy as np
import os
from PIL import Image

# Function to add noise to an image
def add_noise(image):
    """Function that adds Gaussian noise to an image"""
    row, col = image.shape  # Get image dimensions
    mean = 0
    sigma = 15  # Noise intensity
    gauss = np.random.normal(mean, sigma, (row, col))
    noisy = image + gauss
    return np.clip(noisy, 0, 255).astype(np.uint8)

# Paths
dataset_path = r"C:/Users/linoyh/ScientificPython/DL/Split_Dataset/train/Cat Images"
corrupted_folder = r"C:/Users/linoyh/ScientificPython/DL/Split_Dataset/Corrupted_Images"

# Create the corrupted images folder if it doesn't exist
os.makedirs(corrupted_folder, exist_ok=True)

# Iterate over all images in the folder
for img_name in os.listdir(dataset_path):
    img_path = os.path.join(dataset_path, img_name)

    # Read the image
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    # If the image can't be loaded, move it to the corrupted images folder
    if img is None:
        print(f"  Warning: Could not read image {img_name}. Moving to corrupted folder.")
        
        # Move the file to the corrupted images folder
        corrupted_path = os.path.join(corrupted_folder, img_name)
        os.rename(img_path, corrupted_path)
        
        print(f" Moved {img_name} to {corrupted_folder}")
    
    else:
        # Add noise and save the image
        noisy_img = add_noise(img)
        noisy_img_path = os.path.join(dataset_path, f"noisy_{img_name}")
        cv2.imwrite(noisy_img_path, noisy_img)
        print(f" Noisy image saved: {noisy_img_path}")

```

```python
#11

import cv2
import numpy as np
import os
from skimage.util import random_noise

# Function to add Gaussian noise
def add_gaussian_noise(image):
    """Add Gaussian noise to an image"""
    mean = 0
    sigma = 15  # Noise standard deviation
    gauss = np.random.normal(mean, sigma, image.shape).astype(np.uint8)
    noisy = cv2.add(image, gauss)
    return noisy

# Function to add Salt & Pepper noise
def add_salt_pepper_noise(image, amount=0.02):
    """Add Salt & Pepper noise to an image"""
    noisy = random_noise(image, mode='s&p', amount=amount)
    return (noisy * 255).astype(np.uint8)

# Paths
dataset_path = r"C:/Users/linoyh/ScientificPython/DL/Split_Dataset/train/Cat Images"
output_folder = r"C:/Users/linoyh/ScientificPython/DL/Noisy_Dataset"
os.makedirs(output_folder, exist_ok=True)

# Process all images
for img_name in os.listdir(dataset_path):
    img_path = os.path.join(dataset_path, img_name)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    if img is not None:
        noisy_gaussian = add_gaussian_noise(img)
        noisy_sp = add_salt_pepper_noise(img)

        cv2.imwrite(os.path.join(output_folder, f"gaussian_{img_name}"), noisy_gaussian)
        cv2.imwrite(os.path.join(output_folder, f"sp_{img_name}"), noisy_sp)
        print(f"Noisy images saved for: {img_name}")
    else:
        print(f" Skipped corrupted image: {img_name}")
```

```python
import cv2
import numpy as np
import os
from skimage.util import random_noise

# Function to add strong Gaussian noise
def add_strong_gaussian_noise(image):
    mean = 0
    sigma = 50  # Increase noise level
    gauss = np.random.normal(mean, sigma, image.shape).astype(np.uint8)
    noisy = cv2.add(image, gauss)
    return noisy

# Function to add strong Salt & Pepper noise
def add_strong_salt_pepper_noise(image, amount=0.2):  # Increase noise amount
    noisy = random_noise(image, mode='s&p', amount=amount)
    return (noisy * 255).astype(np.uint8)

# Function to reduce resolution to 8x8 and upscale back to 64x64
def degrade_resolution(image):
    low_res = cv2.resize(image, (8, 8), interpolation=cv2.INTER_NEAREST)
    high_res = cv2.resize(low_res, (64, 64), interpolation=cv2.INTER_NEAREST)
    return high_res

# Paths
dataset_path = r"C:/Users/linoyh/ScientificPython/DL/Split_Dataset/train"
output_folder = r"C:/Users/linoyh/ScientificPython/DL/Degraded_Dataset"
os.makedirs(output_folder, exist_ok=True)

# Process all images
for category in os.listdir(dataset_path):
    category_path = os.path.join(dataset_path, category)
    output_category_path = os.path.join(output_folder, category)
    os.makedirs(output_category_path, exist_ok=True)

    for img_name in os.listdir(category_path):
        img_path = os.path.join(category_path, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

        if img is not None:
            img = cv2.resize(img, (64, 64))  # Ensure size

            # Apply severe transformations
            noisy_gaussian = add_strong_gaussian_noise(img)
            noisy_sp = add_strong_salt_pepper_noise(img)
            degraded = degrade_resolution(img)

            # Save transformed images
            cv2.imwrite(os.path.join(output_category_path, f"gaussian_{img_name}"), noisy_gaussian)
            cv2.imwrite(os.path.join(output_category_path, f"sp_{img_name}"), noisy_sp)
            cv2.imwrite(os.path.join(output_category_path, f"degraded_{img_name}"), degraded)

            print(f"✔ Noisy & degraded images saved for: {img_name}")
        else:
            print(f"⚠ Skipped corrupted image: {img_name}")
```
