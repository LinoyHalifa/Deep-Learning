```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

x_range = np.linspace(0, 10, 100)  
w1, w2, w3, w4 = np.random.rand(4)  


x2, x3, x4 = np.random.rand(3)

Y = sigmoid(w1 * x_range + w2 * x2 + w3 * x3 + w4 * x4)

Y_derivative = sigmoid_derivative(w1 * x_range + w2 * x2 + w3 * x3 + w4 * x4) * w1

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_range, Y, label='Y = sigmoid(w1*x1 + w2*x2 + w3*x3 + w4*x4)')
plt.xlabel('x1')
plt.ylabel('Y')
plt.title('Y as a function of x1')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_range, Y_derivative, label="Derivative of Y w.r.t x1")
plt.xlabel('x1')
plt.ylabel("dY/dx1")
plt.title('Derivative of Y as a function of x1')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def Tahn(x):
    return (np.exp(x) - np.exp(-x))  / (np.exp(x) + np.exp(-x))

def Tahn_derivative(x):
    return 1 - Tahn(x) ** 2

x_range = np.linspace(0, 10, 100)  
w1, w2, w3, w4 = np.random.rand(4)  


x2, x3, x4 = np.random.rand(3)

Y = Tahn(w1 * x_range + w2 * x2 + w3 * x3 + w4 * x4)

Y_derivative = Tahn_derivative(w1 * x_range + w2 * x2 + w3 * x3 + w4 * x4) * w1

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_range, Y, label='Y = Tahn(w1*x1 + w2*x2 + w3*x3 + w4*x4)')
plt.xlabel('x1')
plt.ylabel('Y')
plt.title('Y as a function of x1')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_range, Y_derivative, label="Derivative of Y w.r.t x1")
plt.xlabel('x1')
plt.ylabel("dY/dx1")
plt.title('Derivative of Y as a function of x1')
plt.legend()

plt.tight_layout()
plt.show()
```
