```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Exploratory Data Analysis
plt.figure(figsize=(15, 10))

# 1. Sample Digit Visualization
plt.subplot(2, 3, 1)
sample_digits = X_train[:25]
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(sample_digits[i], cmap='gray')
    plt.axis('off')
plt.suptitle('Sample MNIST Digits')

# 2. Class Distribution
plt.subplot(2, 3, 4)
digit_counts = np.bincount(y_train)
plt.bar(range(10), digit_counts)
plt.title('Class Distribution')
plt.xlabel('Digit')
plt.ylabel('Count')

# 3. Pixel Intensity Histogram
plt.subplot(2, 3, 5)
plt.hist(X_train.ravel(), bins=50)
plt.title('Pixel Intensity Distribution')

# 4. Mean Digit Images
plt.subplot(2, 3, 6)
mean_digits = [np.mean(X_train[y_train == i], axis=0) for i in range(10)]
for i, digit in enumerate(mean_digits):
    plt.subplot(2, 5, i+1)
    plt.imshow(digit, cmap='gray')
    plt.title(f'Mean Digit {i}')
    plt.axis('off')

plt.tight_layout()

# Preprocessing
X_train_reshaped = X_train.reshape(-1, 28*28) / 255.0
X_test_reshaped = X_test.reshape(-1, 28*28) / 255.0

# One-hot encode labels
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)

# Neural Network Model
model = Sequential([
    Flatten(input_shape=(28*28,)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) #build the neuron network
model.fit(X_train_reshaped, y_train_encoded, epochs=5, batch_size=32, validation_split=0.2, verbose=0)

# Predictions
y_pred = model.predict(X_test_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = y_test

# Confusion Matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true_classes, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

# Performance Report
print("\nClassification Report:")
print(classification_report(y_true_classes, y_pred_classes))

plt.show()
```
