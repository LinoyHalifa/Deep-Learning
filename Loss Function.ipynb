```python
import numpy as np
import matplotlib.pyplot as plt
import math

class LossFunctions:
    @staticmethod
    def mean_squared_error(y_true, y_pred): # MSE function : (y_true - y_pred) ** 2
        """
        Mean Squared Error (MSE) Loss
        Primarily used for regression problems
        """
        return np.mean((y_true - y_pred) ** 2)

    @staticmethod
    def mean_absolute_error(y_true, y_pred): # absolute value of |y_true - y_pred|
        """
        Mean Absolute Error (MAE) Loss
        Less sensitive to outliers compared to MSE
        """
        return np.mean(np.abs(y_true - y_pred))

    @staticmethod
    def binary_cross_entropy(y_true, y_pred):
        """
        Binary Cross-Entropy Loss
        Used for binary classification problems
        """
        # Prevent log(0)
        epsilon = 1e-15
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        return -np.mean(
            y_true * np.log(y_pred) +
            (1 - y_true) * np.log(1 - y_pred)
        )

    @staticmethod
    def categorical_cross_entropy(y_true, y_pred):
        """
        Categorical Cross-Entropy Loss
        Used for multi-class classification
        """
        epsilon = 1e-15
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        return -np.sum(y_true * np.log(y_pred))

    @staticmethod
    def huber_loss(y_true, y_pred, delta=1.0):
        """
        Huber Loss
        Combination of MSE and MAE
        Less sensitive to outliers
        """
        error = y_true - y_pred
        is_small_error = np.abs(error) <= delta
        squared_loss = 0.5 * error ** 2
        linear_loss = delta * (np.abs(error) - 0.5 * delta)
        return np.mean(np.where(is_small_error, squared_loss, linear_loss))

def visualize_loss_functions():
    """
    Visualize different loss functions
    """
    plt.figure(figsize=(15, 10))

    # MSE Visualization
    plt.subplot(2, 3, 1)
    y_true = np.array([1])
    y_pred = np.linspace(0, 2, 100)
    mse_losses = [LossFunctions.mean_squared_error(y_true, np.array([p])) for p in y_pred]
    plt.plot(y_pred, mse_losses)
    plt.title('Mean Squared Error (MSE)')
    plt.xlabel('Prediction')
    plt.ylabel('Loss')

    # MAE Visualization
    plt.subplot(2, 3, 2)
    mae_losses = [LossFunctions.mean_absolute_error(y_true, np.array([p])) for p in y_pred]
    plt.plot(y_pred, mae_losses)
    plt.title('Mean Absolute Error (MAE)')
    plt.xlabel('Prediction')
    plt.ylabel('Loss')

    # Binary Cross-Entropy Visualization
    plt.subplot(2, 3, 3)
    y_true_binary = 1
    y_pred_binary = np.linspace(0.01, 0.99, 100)
    bce_losses = [LossFunctions.binary_cross_entropy(np.array([y_true_binary]), np.array([p])) for p in y_pred_binary]
    plt.plot(y_pred_binary, bce_losses)
    plt.title('Binary Cross-Entropy')
    plt.xlabel('Prediction Probability')
    plt.ylabel('Loss')

    # Huber Loss Visualization
    plt.subplot(2, 3, 4)
    huber_losses = [LossFunctions.huber_loss(y_true, np.array([p])) for p in y_pred]
    plt.plot(y_pred, huber_losses)
    plt.title('Huber Loss')
    plt.xlabel('Prediction')
    plt.ylabel('Loss')

    # Categorical Cross-Entropy Visualization
    plt.subplot(2, 3, 5)
    y_true_cat = np.array([1, 0, 0])
    y_pred_cat = np.linspace(0.01, 0.99, 100)
    cce_losses = []
    for p in y_pred_cat:
        pred = np.array([p, 1 - p / 2, 1 - p / 2])
        cce_losses.append(LossFunctions.categorical_cross_entropy(y_true_cat, pred))
    plt.plot(y_pred_cat, cce_losses)
    plt.title('Categorical Cross-Entropy')
    plt.xlabel('Prediction Probability')
    plt.ylabel('Loss')

    plt.tight_layout()
    plt.show()

def demonstrate_loss_functions():
    """
    Demonstrate different loss function behaviors
    """
    # Regression Example
    print("Regression Loss Functions:")
    y_true_reg = np.array([10])
    y_pred_reg_scenarios = [8, 10, 12, 15]

    for pred in y_pred_reg_scenarios:
        print(f"\nPrediction: {pred}")
        print(f"MSE Loss: {LossFunctions.mean_squared_error(y_true_reg, np.array([pred])):.4f}")
        print(f"MAE Loss: {LossFunctions.mean_absolute_error(y_true_reg, np.array([pred])):.4f}")
        print(f"Huber Loss: {LossFunctions.huber_loss(y_true_reg, np.array([pred])):.4f}")

    # Classification Example
    print("\n\nClassification Loss Functions:")
    y_true_binary = np.array([1])
    y_pred_binary_scenarios = [0.1, 0.3, 0.5, 0.7, 0.9]

    for pred in y_pred_binary_scenarios:
        print(f"\nPrediction Probability: {pred}")
        print(f"Binary Cross-Entropy Loss: {LossFunctions.binary_cross_entropy(y_true_binary, np.array([pred])):.4f}")

    # Multiclass Example
    print("\n\nMulticlass Classification:")
    y_true_multi = np.array([1, 0, 0])  # One-hot encoded
    y_pred_multi_scenarios = [
        np.array([0.9, 0.1, 0.0]),
        np.array([0.7, 0.2, 0.1]),
        np.array([0.5, 0.3, 0.2])
    ]

    for pred in y_pred_multi_scenarios:
        print(f"\nPrediction: {pred}")
        print(f"Categorical Cross-Entropy Loss: {LossFunctions.categorical_cross_entropy(y_true_multi, pred):.4f}")


def main():
    # Demonstrate loss function calculations
    demonstrate_loss_functions()

    # Visualize loss functions
    visualize_loss_functions()

if __name__ == "__main__":
    main()
```
