```python
# Libraries for this section

from keras.layers import Dense, SimpleRNN
from keras.models import Sequential # in this situation we don't have skip connection
import altair as alt
import pandas as pd
import numpy as np

# features
X = np.array([[[0, 0, 1, 1]],
              [[0, 1, 0, 1]]]).T # here we have 2 vetors of inputs. we make to the X vector transpose

# expected values
y = np.array([[0, 1, 1, 0]]).T

print(f'training data shape: {X.shape}')
print(f'targets data shape: {y.shape}')

# Define a network as a linear stack of layers
model = Sequential() # open a Sequential model

# Add a recurrent layer with 2 units
model.add(SimpleRNN(2, input_shape=(1, 2))) # in this case we add just one layer. this layer gets 1 row and 2 columns of input

# Add the output layer with a sigmoid activation
model.add(Dense(1, activation='tanh')) # output layer with hyperbolic tanh

model.summary() # presents our network and how many layers and the size of layers

model.compile(optimizer='Adadelta', # make compilation to the model. it build the network and on each arch il build the forward. this compile 
              loss='mean_squared_error', #plan the network and the structure of the archs. optimazer is the backpropagation. so the optimazer
              metrics=['acc']) # tell us in which way to make the backpropagation. in the marics we define how to test the quality of the result 
# metrics - confusion metrix
history = model.fit(X, y, # here we train the model. we make here the train with 5000 epochs
                    epochs=5000,
                    verbose=0)
alt.data_transformers.disable_max_rows()

loss = history.history['loss']
accuracy = history.history['acc']

df = pd.DataFrame({"accuracy":accuracy, "loss":loss, "time-step": np.arange(0, len(accuracy))})

base = alt.Chart(df).mark_line(color="blue").encode(x="time-step", y="accuracy")
loss = alt.Chart(df).mark_line(color="red").encode(x="time-step", y="loss")
((base  + loss).properties(title='Chart 2').resolve_scale(y='independent'))

# here in this example we make between h and output we use dense with 3 parameters.
# training data shape: (4, 1, 2) - size of X
# in the summary : we have one layer of simple RNN and second layer of dense.
# to the simple RNN we have 2 entries and in the dense we have 1 entry
#param- tell us how many Weights and biases we have. in the RNN we have 10 and in the dense we have 3 
```
